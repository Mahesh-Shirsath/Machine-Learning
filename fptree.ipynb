```python
import pyfpgrowth
```

```python
transactions = [['a', 'b', 'e'],
                ['b', 'd'],
                ['b', 'c'],
                ['a', 'b', 'd'],
                ['a', 'c'],
                ['b', 'c'],
                ['a', 'c'],
                ['a', 'b', 'c', 'e'],
                ['a', 'b', 'c']]
```

```python
patterns = pyfpgrowth.find_frequent_patterns(transactions, 4)
```

Now we find patterns that are associated with another with a certain minimum probability

```python
rules = pyfpgrowth.generate_association_rules(patterns, 0.7)
```
