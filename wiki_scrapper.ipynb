```python
# !pip install request bs4

from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/Web_scraping"
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
```

```python
print(soup.title.get_text())
```

```python
print(soup.find("h1", id="firstHeading").get_text())
```

```python
short_summary = soup.find("div", id="bodyContent").find("div", class_="mw-parser-output").find("p").get_text()
print(short_summary)
```

```python
summary = "\n".join(reversed([ x.get_text() for x in soup.find("div", id="toc").find_all_previous("p") ]))
print(summary)
```
